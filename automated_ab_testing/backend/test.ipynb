### Full prompt

```python
prompt = generate_prompt(html_content, user_input="We are a luxury brand with more mature clients.")
print(prompt)
```

### Generation

```python
messages = [{"role": "user", "content": prompt}]
print(messages)
```

```python
response = completion(
        messages=messages,
        model="o1-mini",
        temperature=1,
        max_tokens=65536,
    )
```

```python
print(response)

new_html = response.choices[0].message.content
print(new_html)
```
